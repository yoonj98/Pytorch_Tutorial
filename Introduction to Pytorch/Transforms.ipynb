### Transforms
FashionMNIST 기능은 PIL 이미지 형식이며 레이블은 정수입니다. 훈련을 위해서는 정규화된 텐서로서의 기능과 원-핫 인코딩된 텐서로서의 레이블이 필요합니다. 이러한 변환을 수행하기 위해 ToTensor 및 Lambda를 사용합니다.

```python
import torch
from torchvision import datasets
from torchvision.transforms import ToTensor, Lambda

ds = datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,
    transform=ToTensor(),
    target_transform=Lambda(lambda y: torch.zeros(10, dtype=torch.float).scatter_(0, torch.tensor(y), value=1))
)
```

### Tensor()
ToTensor는 PIL 이미지 또는 NumPy ndarray를 FloatTensor. [0., 1.] 범위에서 이미지의 픽셀 강도 값을 조정  
> 다음 예제에서는 사용자 변환 함수인 `lambda()`를 통해 크기 10의 0 텐서(데이터 세트의 레이블 수)를 생성하고 `scatter_`를 호출하여 레이블 y에 주어진 대로 인덱스에 값=1을 할당한다.

```python
target_transform = Lambda(lambda y: torch.zeros(
    10, dtype=torch.float).scatter_(dim=0, index=torch.tensor(y), value=1))
```
