### Tensors
PyTorch에서는 텐서를 사용하여 모델의 입력 및 출력과 모델의 매개 변수를 인코딩한다.

```python
import torch
import numpy as np
```

#### Tensor Initialize

```python
data = [[1, 2],[3, 4]]
x_data = torch.tensor(data)
```

1. Numpy array로부터 생성

```python
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
```

2. 다른 tensor로부터 생성  
텐서는 명시적으로 재정의하지 않는 한 인수 텐서의 속성(모양, 데이터 유형)을 유지

```python
x_ones = torch.ones_like(x_data) # retains the properties of x_data
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype=torch.float) # overrides the datatype of x_data
print(f"Random Tensor: \n {x_rand} \n")
```

3. random 상수 및 고정 상수로부터 생성  
`shape` : tensor 차원의 튜플

```python
shape = (2,3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor}")
```

### Operations on Tensors
기본적으로 텐서는 CPU에서 생성됩니다. .to메서드를 사용하여 텐서를 GPU로 명시적으로 이동해야 합니다 (GPU 가용성 확인 후)

```python
# gpu 가용성 확인
if torch.cuda.is_available():
    tensor = tensor.to('cuda')
```

인덱싱 및 슬라이싱 : numpy와 동일

```python
tensor = torch.ones(4, 4)
print('First row: ', tensor[0])
print('First column: ', tensor[:, 0])
print('Last column:', tensor[..., -1])
tensor[:,1] = 0
print(tensor)
```

Tensor 결합  
`torch.cat` : 주어진 차원을 따라 텐서 시퀀스를 연결

```python
t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)
```

산술연산

```python
y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)

y3 = torch.rand_like(tensor)
torch.matmul(tensor, tensor.T, out=y3)

z1 = tensor * tensor
z2 = tensor.mul(tensor)

z3 = torch.rand_like(tensor)
torch.mul(tensor, tensor, out=z3)
```

단일 tensor가 있는 경우 `item()`을 통해 변환 가능

```python
agg = tensor.sum()
agg_item = agg.item()
print(agg_item, type(agg_item))
```

In-place operations  
결과를 피연산자에 저장

```python
print(tensor, "\n")
tensor.add_(5)
print(tensor)
```

### Bridge with NumPy  
CPU 및 NumPy 어레이의 텐서는 기본 메모리 위치를 공유할 수 있으며 하나를 변경하면 다른 하나가 변경됨.

Tensor to Numpy array

```python
t = torch.ones(5)
print(f"t: {t}")
n = t.numpy()
print(f"n: {n}")
```

```python
t.add_(1)
print(f"t: {t}")
print(f"n: {n}")
```

Numpy array to tensor

```python
n = np.ones(5)
t = torch.from_numpy(n)
```

```python
np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
```
